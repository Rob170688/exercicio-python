```python
senha_correta = "1234"

while True:
    senha = input("Digite uma senha com 4 digitos: ")
    if senha == senha_correta:
        print("Senha correta! Acesso concedido.")
        break
    else:
        print("Senha incorreta! Tente novamente.")
```

```python
total_produto = 0

while True:
  produto = input("Digite o valor do produto (ou um número negativo para finalizar): ")

  if float(produto) > 0:
   total_produto += float(produto)

  else:
   print(f"O valor total da compra é: R$ {total_produto:.2f}")
   break
```

```python
numero = 0
opcao = 0

numero = input("Digite um número inteiro: ")

while True:
    opcao = input(f"Se você quiser converter {numero} para Binário digite '1' ou se você quiser converter para Hexadecimal digite '2'.")

    if opcao == 1:
       print(f"O número {numero} em binário é {bin(numero)[2:]}")

    elif opcao == 2:
       print(f"O número {numero} em hexadecimal é {hex(numero)[2:]}")

    elif opcao < 0:
       break
```
