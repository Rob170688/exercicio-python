```python
num1 = int(input("Escreva um numero: "))
num2 = int(input("Escreva mais um numero: "))
soma = num1 + num2
subtra = num2 - num1
prod = num1 * num2
divis = num1 / num2
print("a. A soma dos numeros é:", soma)
print("b. A subtração do segundo numero pelo primeiro é:", subtra)
print("c. O produto dos dois numeros é:", prod)
print("d. A divisão do primeiro numero pelo segundo é:", divis)
```
