```python
idade = int(input('Digite sua idade: '))
if idade >= 10 and idade < 20:
    print('Você é Adolescente')
elif idade >= 20 and idade < 30:
    print('Você é Jovem')
elif idade >= 30 and idade <= 100:
    print('Você é Adulto')
else:
    print('Você está fazendo hora Extra!')
```

```python
num1 = int(input('Digite o primeiro número: '))
num2 = int(input('Digite o segundo número: '))
if num1 < num2:
  print('O número maior é', num2)
elif num1 > num2:
  print('O número maior é', num1)
else:
  print('Coloque números diferentes.')
```

```python
valor = int(input('Digite um valor: '))
if valor < 0:
  print('Seu valor é Negativo.')
elif valor > 0:
  print('Seu valor é Positivo.')
else:
  print('Sem valor.')
```

```python
sexo = input('Digite F se você for Mulher ou M se você for Homem.')
if sexo == 'F':
  print('Mulher')
elif sexo == 'M':
  print('Homem')
else:
  print('Sexo Inválido')
```
